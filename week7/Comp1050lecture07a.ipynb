Run first one of the two functions `cf_color`. One is an incorrect version, the other a corrected version. 
At the bottom you will find code that tests the implementation. These tests were provided by students during the lecture.

Note, the code used during the lecture had a mistake, that was found because of test input provided by the students.

```python
def cf_color(frogs, clutch):
    """Incorrect version. The indentation needs to be fixed."""
    cfc = "white"
    if clutch >= -2 and clutch<=2:
        if frogs >= 0 and frogs <= 2:
            cfc = "blue"
        if clutch >= 0:
            cfc = "green"
    if clutch >= frogs:
        cfc = "yellow"
    elif frogs <= 0 and frogs >= -2:
        cfc = "black"
    if clutch >= 0:
        cfc = "orange"
    elif frogs + clutch >= -2:
        cfc = "red"
    return cfc
```

```python
def cf_color(frogs, clutch):
    """Corrected version. Reference implmentation."""
    cfc = "white"
    if clutch >= -2 and clutch<=2:
        if frogs >= 0 and frogs <= 2:
            cfc = "blue"
            if clutch >= 0:
                cfc = "green"
            if clutch >= frogs:
                cfc = "yellow"
        elif frogs <= 0 and frogs >= -2:
            cfc = "black"
            if clutch >= 0:
                cfc = "orange"
            elif frogs + clutch >= -2:
                cfc = "red"
    return cfc
```
