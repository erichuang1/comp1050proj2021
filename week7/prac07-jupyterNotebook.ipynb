```python
#################################################
# "broken" function 
#################################################
# This function takes in three parameters
#    day (e.g. 1, 2, 3, ...) for days of the month 
#    month (e.g. 1, 2, 3, ...) for (Jan, Feb, March, ...)
#    year (e.g. 1999, 2001, 2089) for years between 1700 and 2100

# This function returns:
#     true if the (day, month, year) make a valid date between 1700 and 2100
#     false in any other case
#     but it is incomplete... use the test data to determine

def isValidDate(day, month, year):
    isValidDate = True
    
    #check if day, month, and year are whoe numbers... if not, then it isn't a valid date (Return false)
    if (type(day) is not int) or (type(month) is not int) or (type(year) is not int):
        isValidDate = False
    
    #check that the year is between 1700 and 2099 
    elif year < 1700 or year > 2099:
        isValidDate = False
        
    elif month < 1 or month > 12:
        isValidDate = False
    
    #You can add your other coding checks here if you like .   <<<-------------
    #When are some other times when this block of code should return false?

    elif day <= 0:
        isValidDate = False

    elif month in [1,3,5,7,8,10,12]:
        if day > 31:
            isValidDate = False

    elif month in [4,6,9,11]:
        if day > 30:
            isValidDate = False
            
    elif month == 2:
        if year % 4 == 0:
            if year % 100 == 0:
                if year % 400 == 0:
                    # leap
                    if day > 29:
                        isValidDate = False
                else: 
                    # not leap
                    if day > 28:
                        isValidDate = False
            else:
                # leap
                if day > 29:
                    isValidDate = False
        else:
            # not leap
            if day > 28:
                isValidDate = False
       
    if year == 1752 and month == 9:
        if day >= 3 and day <= 13:
            isValidDate = False

                
    # Step 1: If the year is evenly divisible by 4, go to step 2. Else, not leap.

    # Step 2: If the year is evenly divisible by 100, go to step 3. Otherwise, is leap.

    # Step 3: If the year is evenly divisible by 400, is leap. Otherwise, not leap.

    #do not modify this line.
    return isValidDate 



#there should be no output from this cell
```
